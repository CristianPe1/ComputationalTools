```python
# global imports
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# function to return the derivatives
def derivatives(y,a): 
    return np.array([-y[1] + a*y[0]*(np.power(y[0],2) + np.power(y[1],2)),
                     y[0] + a*y[1]*(np.power(y[0],2) + np.power(y[1],2))])
```

```python
%matplotlib inline
def graph(a):
    sns.set()
    sns.set_context('poster')

    N = 1000 # Number of initial conditions
    LIM =10.0; 
    X, Y = np.meshgrid(np.linspace(-LIM, LIM, N), np.linspace(-LIM, LIM, N))
    DX, DY = derivatives([X, Y],a)

    M = 1000
    E = M*9.81*X + 0.5*M*Y*Y

    fig = plt.figure(figsize=(10,10))
    plt.streamplot(X, Y, DX, DY, density=[1, 1], color=E, arrowsize=4, arrowstyle='fancy', linewidth=2, cmap='plasma')
    plt.colorbar()

def graph_quiver(a):
    sns.set()
    sns.set_context('poster')

    N = 1000 # Number of initial conditions
    LIM =10.0; 
    X, Y = np.meshgrid(np.linspace(-LIM, LIM, N), np.linspace(-LIM, LIM, N))
    DX, DY = derivatives([X, Y],a)

    M = 1000
    E = M*9.81*X + 0.5*M*Y*Y

    fig, ax = plt.subplots(figsize =(9, 9))
    ax.quiver(X, Y, DX, DY)
 
    ax.xaxis.set_ticks([])
    ax.yaxis.set_ticks([])
    ax.set_aspect('equal')
 
    # show plot
    plt.show()
```

```python
graph(-0.01)
```

```python
graph_quiver(-0.01)
```

```python
graph(0)
```

```python
graph_quiver(0)
```
