El algoritmo de Gram - Schimidt consiste en seguir la siguiente ecuaciÃ³n:
$$
u_k = v_k - \sum_{j=1}^{k}\frac{\langle v_k,u_j \rangle}{\langle u_j,u_j \rangle}u_j $$

En el espacio de Hilbert los vectores son funciones y el producto interno entre dos de estas esta definido como:
$$
\langle f_1(x),f_2(x)\rangle =\int_{a}^{b} f_1(x)f_2^*(x)p(x)dx
$$

```python
"""def ProductoInterno(f1,f2,p,a,b):

    arg = lambda x : f1(x)*f2(x)*p(x)

    I = quad(arg, a, b)

    return I"""

def ProductoInterno(f1,f2,p,a,b):

    arg = f1*f2*p

    I = sp.integrate(arg, (x, a, b))

    return I


def Correcion(vk,u,PI):
    #print(f"uc: {u}")
    s=np.float128(0)
    for uj in u:
        #print(vk,uj)
        PI1 = PI(vk,uj)
        #print(f"PI1: {PI1}")
        PI2 = PI(uj,uj)
        #print(f"PI2: {PI2}")
        if PI2 == 0:
            s +=0
        else:
            s += sp.Rational(PI1,PI2)*uj 
        #print(f"vk,uj,s: {vk,uj,s}")
    return s

def GramSchmidt(v,PI):
    
    u = np.zeros(len(v))
    
    x = sp.Symbol('x')

    v = sp.Matrix(v)
    u = sp.Matrix(u)

    for k in range(len(v)):
        #print(f"ubc:{u[k]}")
        u[k] = v[k] - Correcion(v[k],u,PI)
        if k>=1: u[k] = u[k]*(1/u[k].subs(x,1))
        print(u[k].subs(x,1))
        #if k>=1: u[k] = u[k]*(1/sp.sqrt(PI(u[k],u[k])))
        #print(PI(u[k],u[k]))
        
    return u
```
