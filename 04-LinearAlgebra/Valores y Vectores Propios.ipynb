```python
a, b, c, d = sp.symbols('a b c d')
M = sp.Matrix([[a,  c], 
              [c,  b]])
M
```

```python
 a, b, c, d = sp.symbols('a b c d')

M = sp.Matrix([[a, 0,  a], 
            [0,  -b, 0], 
            [a, 0,  a]])
M
```
